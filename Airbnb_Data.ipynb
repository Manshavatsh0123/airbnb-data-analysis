```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data.csv")
```

```python
df.head()
```

```python
df.columns
```

# Checking missing values

```python
print(df.isnull().sum()) 
```

```python
df.info() # Type of data in columns
```

# Handling missing values 

```python
df['last review'] = pd.to_datetime(df['last review'],errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0, 'last review':df['last review'].min()},inplace=True)
```

```python
df.dropna(subset =['NAME','host name'],inplace=True)
```

```python
print(df.isnull().sum())
```

# Drop columns

```python
df = df.drop(columns=['license','house_rules'],errors='ignore')
```

```python
df.head()
```

# Remove dollars signs and convert to float

```python
df['price'] = df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

# remove duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

# descriptive statistics

```python
df.describe()
```

# Data visualization

# What is distribution of listing prices?

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'], bins=50 ,kde=True,color= 'red')

plt.title("Distribution of listing price")
plt.xlabel("Price $")
plt.ylabel("Frequency")

plt.savefig("Graph/listing_prices.png", dpi=300, bbox_inches="tight")
```

```python
# The histogram shows a fairly even distribution of listing prices across different price ranges, indicating no particular concentration of listings in any specific price range. The KDE line helps visualize this even spread more clearly, confirming that the dataset contains listings with a wide variety of prices.
```

# How different room type distributed?

```python
plt.figure(figsize = (8,6))
sns.countplot(x='room type',data=df,color='hotpink')

plt.title("Room types")
plt.xlabel("Room type")
plt.ylabel("count")

plt.savefig("Graph/room_type.png", dpi=300, bbox_inches="tight")
```

# How are listing distributed across different neighborhoods?

```python
plt.figure(figsize = (12,8))
sns.countplot(x='neighbourhood group',data=df,color='lightgreen',order=df['neighbourhood group'].value_counts().index)

plt.title("Number of listing by neighbourhood group")
plt.xlabel("count")
plt.ylabel("Neighbourhood group")

plt.savefig("Graph/neighborhood.png", dpi=300, bbox_inches="tight")
plt.show()
```

# what is relationship between price and room type ?

```python
plt.figure(figsize = (10,6))
sns.boxplot(x='room type',y="price",data=df,hue='room type',palette='Set1')

plt.title("Price vs Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price ($)")
plt.legend(title='Room type')

plt.savefig("Graph/price_over_roomtype.png", dpi=300, bbox_inches="tight")
plt.show()
```

# How has the number of reviews change over with time

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line', color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')

plt.savefig("Graph/reviews_over_time.png", dpi=300, bbox_inches="tight")
plt.show()
```
